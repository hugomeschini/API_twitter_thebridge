### Pandas Dataframe to SQL

```python
import sqlite3

conn = sqlite3.connect('df_database')
c = conn.cursor()
```

```python
df_obj = df[['Date', 'User', 'Tweet']]
df_int = df[['Likes', 'Retweets', 'Replies', 'id']]
```

```python
c.execute('CREATE TABLE IF NOT EXISTS products (Date, User, Tweet)')
conn.commit()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS products (Likes, Retweets, Replies, id)')
conn.commit()
```

```python
df_obj.to_sql('str_data', conn, if_exists='replace', index = False)

df_int.to_sql('int_data', conn, if_exists='replace', index = False)
```

```python
c.execute('''  
SELECT * FROM int_data
          ''')

for row in c.fetchall():
    print (row)
```
